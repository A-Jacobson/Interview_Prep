```python
%matplotlib inline
import matplotlib.pyplot as plt
from ipywidgets import widgets
from IPython import display
import numpy as np
import seaborn as sns
import math
from ipywidgets import interact
import pandas as pd
```

Big-O notation is used to track the efficiency of algorithms. Big-O measures the (worst case) number of steps the algorithm will take to solve a problem depending on the number of inputs it recieves. 

```python
def plot_big_o(n):
    number_inputs = np.arange(1,n+1)
    bigo = pd.DataFrame(number_inputs, columns=['n'])
    bigo['logarithmic'] = np.log2(bigo['n'])
    bigo['linear'] = bigo['n']
    bigo['log linear'] = np.log2(bigo['n']) * bigo['n']
    bigo['quadratic'] = bigo['n'] ** 2
#     bigo['cubic'] = bigo['n'] ** 3
#     bigo['exponential'] = (np.ones_like(number_inputs)*2) ** bigo['n']
#     bigo['factorial'] = bigo['n'].map(math.factorial)
    del bigo['n']
    bigo.plot()
    plt.ylabel("Number of Operations")
    plt.xlabel("Number of Inputs")
    plt.show()
```

```python
interact(plot_big_o, n=(1, 201));
```

### BigO of python data structures
### Lists
|     Operation    | Big-O    |
|:----------------:|----------|
| index[]          | O(1)     |
| index assignment | O(1)     |
| append           | O(1)     |
| pop()            | O(1)     |
| pop(i)           | O(n)     |
| insert(i, item)  | O(n)     |
| contains (in)    | O(n)     |
| get slice [x:y]  | O(k)     |
| del slice        | O(n)     |
| reverse          | O(n)     |
| concatenate      | O(k)     |
| sort             | O(nlogn) |
| multiply         |   O(nk)  |

### Dictionaries

|     Operation    | Big-O    |
|:----------------:|----------|
| copy[]           | O(1)     |
| get item         | O(1)     |
| set item         | O(1)     |
| delete item      | O(1)     |
| contains (in)    | O(1)     |
| iteration        | O(n)     |
